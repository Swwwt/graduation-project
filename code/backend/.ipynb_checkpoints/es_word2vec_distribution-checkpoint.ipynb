```python
import csv

scores = []
distances = []
with open('./search_result_summary.csv', newline='') as csvfile:
    
    reader = csv.DictReader(csvfile)#, quoting = csv.QUOTE_NONE)
    for row in reader:
        #print(row['id'], row['score'], row['distance'])
        scores.append(row['score'])
        distances.append(row['distance'])
```

```python
scores = [float(i) for i in scores if i!='']
print(scores)
```
